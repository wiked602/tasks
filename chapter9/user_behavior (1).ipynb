```python
labels = ['Android','IOS']
values = ['79','21']
plt.pie(values,labels=labels,autopct='%1.1f%%')
plt.title("Процент использования Android и IOS")
plt.show()
```

```python
df['User Behavior Class'].value_counts()
plt.figure(figsize = (10,5))
clas = [1,2,3,4,5]
cnt = [136,146,143,139,136]
plt.bar(clas,cnt,label = "Оценка использования смартфонов")
plt.xlabel('Оценка')
plt.ylabel('Кол-во пользователей')
plt.legend()
plt.show()
```

```python
df['Gender'].value_counts()
labels = ['Male','Female']
values = ['52','48']
plt.pie(values,labels=labels,autopct='%1.1f%%')
plt.title("Процент мужчин и женщин среди опрошенных")
plt.show()
```

```python
# Кто больше использует телефон женщины или мужчины
usage = []
gender = []
male = []
female = []
for i in range(700):
    usage.append(df['Screen On Time (hours/day)'][i])
    gender.append(df['Gender'][i])
    if gender[i] == "Male":
        male.append(usage[i])
    else:
        female.append(usage[i])
male_usage = sum(male) // len(male)
female_usage = sum(female) / len(male)
print(male_usage,female_usage) # Следовательно в среднем мужчины пользуются телефоном больше чем женщины,скорее всего это обусловлено тем что среди опрошенных мужчин больше
```
